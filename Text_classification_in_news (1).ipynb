```python
import pandas as pd
data=pd.read_csv('test.csv')
data.head()
```

```python
# Assuming your dataset has 'title', 'description', and 'category' columns
X = data['Title'] + ' ' + data['Description']
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
# Create a TF-IDF vectorizer to convert text data into numerical features
tfidf_vectorizer = TfidfVectorizer(max_features=1000)
```

```python
# Transform the data into TF-IDF features
X_tfidf = tfidf_vectorizer.fit_transform(X)
```

```python
# Apply Latent Dirichlet Allocation (LDA) for topic modeling
num_topics = 5
lda = LatentDirichletAllocation(n_components=num_topics, random_state=42)
topics = lda.fit_transform(X_tfidf)
```

```python
# Assign articles to topics based on the highest topic probability
predicted_categories = topics.argmax(axis=1)
```

```python
# Add the predicted categories to the DataFrame (optional)
data['predicted_category'] = predicted_categories
```

```python
from nltk.corpus import stopwords
import string

punctuation = string.punctuation
data['Title'] = data['Title'].apply(lambda x: ''.join([i for i in x if i not in punctuation]).split())
data['Description'] = data['Description'].apply(lambda x: ''.join([i for i in x if i not in punctuation]).split())
data.head()
```

```python
# Map topics to predefined categories
topic_to_category = {
    0: "Technology",
    1: "Sports",
    2: "Politics",
    3: "Entertainment",
    4: "Health"
}
```

```python
# Assuming 'predicted_topics' is an array of assigned topic numbers for each article
predicted_topics = [0, 1, 2, 3, 4]
```

```python
# Error-checking: Ensure all predicted_topics are valid topic numbers
for topic in predicted_topics:
    if topic not in topic_labels:
        print(f"Invalid topic number: {topic}")
```

```python
# Classify articles based on assigned topics
predicted_categories = [topic_to_category[topic] for topic in predicted_topics]
```
